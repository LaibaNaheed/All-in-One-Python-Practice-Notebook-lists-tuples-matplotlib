QUESTION 1:Creating and Accessing Tuples
1. Create a tuple colors = (&quot;red&quot;, &quot;green&quot;, &quot;blue&quot;, &quot;yellow&quot;).
2.   Print the first, last, and second-to-last elements using indexing.
3.   Print the first 3 elements using slicing.

```python
# Creating a tuple
colors = ("red", "green", "blue", "yellow")

# Accessing elements with indexing
print("First element:", colors[0])        # index 0 → first
print("Last element:", colors[-1])       # index -1 → last
print("Second-to-last element:", colors[-2])  # index -2 → second-to-last

# Accessing elements with slicing
print("First 3 elements:", colors[:3])   # from index 0 up to (but not including) index 3
```

QUESTION 4: Iterating Over Tuples
1. Create a tuple numbers = (10, 20, 30, 40, 50).
2. Use a for loop to print each element.
3. Find the sum and average of elements using iteration.


```python
# 1. Create a tuple
numbers = (10, 20, 30, 40, 50)

# 2. Use a for loop to print each element
print("Tuple elements:")
for num in numbers:
    print(num)

# 3. Find the sum and average using iteration
total = 0
for num in numbers:
    total += num

average = total / len(numbers)

print("Sum of elements:", total)
print("Average of elements:", average)
```

Pie Chart
QUESTION 1:
1. A survey was conducted among students to know their preferred programming languages. The results were:
o Python: 40 students
o Java: 25 students
o C++: 20 students
o JavaScript: 15 students
2. Draw a pie chart showing the popularity of these programming languages.
3. Add labels, percentages, and a suitable title.

```python
import matplotlib.pyplot as plt

def main():
    # Data
    languages = ["Python", "Java", "C++", "JavaScript"]
    students = [40, 25, 20, 15]

    # Pie chart
    plt.pie(students,
            labels=languages,          # Add labels
            autopct='%1.1f%%',         # Show percentages
            startangle=140,            # Rotate chart for better view
            colors=["#66b3ff", "#99ff99", "#ffcc99", "#ff9999"])  # Custom colors

    # Title
    plt.title("Popularity of Programming Languages among Students")

    # Show chart
    plt.show()

main()
```

Histogram
QUESTION 2:
1. A class has students with the following ages:
18, 19, 20, 20, 21, 22, 22, 22, 23, 24, 25, 25, 25, 26, 27
2. Create a histogram to show the distribution of student ages.
3. Use 5 bins, and label the axes and title properly.

```python
import matplotlib.pyplot as plt

# Data: student ages
ages = [18, 19, 20, 20, 21, 22, 22, 22, 23, 24,
        25, 25, 25, 26, 27]

# Create histogram with 5 bins
plt.hist(ages, bins=5, color="lightgreen", edgecolor="black")

# Labels and title
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.title("Distribution of Student Ages")

# Show the histogram
plt.show()
```

Multiple Line Plots
QUESTION 3:
1.  Plot two lines on the same graph:
o First line for even numbers: y = 2x
o Second line for odd numbers: y = 2x - 1
o Use x = [1, 2, 3, 4, 5].
2.  Add different colors, markers, a legend, and a proper title.

```python
import matplotlib.pyplot as plt

# X values
x = [1, 2, 3, 4, 5]

# Equations
y_even = [2 * i for i in x]       # y = 2x
y_odd = [2 * i - 1 for i in x]    # y = 2x - 1

# Plot first line (even numbers)
plt.plot(x, y_even, color="blue", marker="o", label="y = 2x")

# Plot second line (odd numbers)
plt.plot(x, y_odd, color="red", marker="s", label="y = 2x - 1")

# Add title and labels
plt.title("Multiple Line Plot: Even vs Odd")
plt.xlabel("X values")
plt.ylabel("Y values")

# Add legend
plt.legend()

# Show graph
plt.show()
```
